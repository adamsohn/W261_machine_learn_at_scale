### Q1 Student Answers:
> __a)__ Source: Async<br>
First Order Condition for Convexity: The maximum or minimum of an unconstrained (loss) function must occur at a point where the first derivative = 0.<br>
Second Order Condition for Convexity: If the second derivatives are positive where the first derivative of the loss function = 0, this location is the minimum. If the second derivatives are negative where the loss function = 0, this location is the maximum.


Source: Async 6.11
> __b)__ <br>
Problem Domain Space - Problem Domain Space is a view of regression model hypotheses (function lines) as a function of input variables (ie. not model parameters). In Problem Domain Space, model parameters are held constant for each regression model hypothesis.<br>
Model Parameter Space - Model Parameter Space is a view of multiple regression model objective (ie. loss) function iso-surfaces plotted against varying values for model parameters (ie. not input values). In Model Parameter Space, model input values are held constant for each regression model. Gradient descent progression towards local/global minima/maxima can be observed in the space as movement (by varying model parameters while keeping input variables constant) towards tighter-bound iso-surfaces.<br>
Practical Difficulties - The best use case for visual solution method leveraging Problem Domain Space / Model Parameter Space is the simplistic case of y = mx + b. As the quantity of variables increases past 3, visualization techniques will cease to clearly convey the clear optimal path to the human eye. Also, computation of iso-surfaces will be challenging with many variables. Also, situations where there is non-convexity or non-linear regression would cause this method to be untenable. <br><br>

> __c)__ Learning rate is the size of the 'steps' a gradient descent loss function will take modifying model parameters in search of an optimum loss. Tradeoffs associated are that steps that are too-small will result in a lengthy computation to reach optima & steps that are too large may advance past an optima (or path towards optima) for a non-optimal local optima or not converge at all.<br><br>

> __d)__ In the context of OLS, a 'closed form solution' is a problem which can be solved using algebraic methods according to the First/Second Order Conditions for Convexity. This becomes more challenging at-scale (ie. beyond linear regression) as not all functions are differentiable. In addition, large matrices are computationally difficult to invert in order to solve. Complicating this is that inverting matrices is not easy to parallelize.

```
# part d - write function to compute loss (FILL IN MISSING CODE BELOW)
def OLSLoss(dataRDD, W):
    """
    Compute mean squared error.
    Args:
        dataRDD - each record is a tuple of (features_array, y)
        W       - (array) model coefficients with bias at index 0
    """
    augmentedData = dataRDD.map(lambda x: (np.append([1.0], x[0]), x[1]))
    ################## YOUR CODE HERE ##################
    loss = None
    loss_sprk = augmentedData.map(lambda x: ((np.sum(x[0]*W) - x[1])**2,1)) \
      .reduce(lambda x,y: (x[0]+y[0], x[1]+y[1]))
    loss = loss_sprk[0]/loss_sprk[1]
    ################## (END) YOUR CODE ##################
    return loss
```

Gradient: $ \nabla{\boldsymbol{\theta}} f(\boldsymbol{\theta}) = \frac{2}{n}\,\sum{i=1}^{n}\left[ \boldsymbol{\theta}^T\cdot\mathbf{x}'_i - y_i\right] \cdot \mathbf{x}'_i $<br>
'the mean of the data weighted by the errors' - The gradient formula can be explained as 2 \* standard deviation (ie. error) of regression \* mean of data. In this respect, the formula accurately reflects the phrase 'the mean of the data weighted by the errors'<br>
> __b)__ The treatment of Gradient would be very similar to OLS Loss. Both the calculation of individual elements (map) and the summation of these elements (reduce) can be parallelized. The division of the sum by the count is done after the reduce step. <br>

```
# part b - function to perform a single GD step
def GDUpdate(dataRDD, W, learningRate = 0.1):
    """
    Perform one OLS gradient descent step/update.
    Args:
        dataRDD - records are tuples of (features_array, y)
        W       - (array) model coefficients with bias at index 0
    Returns:
        new_model - (array) updated coefficients, bias at index 0
    """
    # add a bias 'feature' of 1 at index 0
    augmentedData = dataRDD.map(lambda x: (np.append([1.0], x[0]), x[1])).cache()
    
    ################## YOUR CODE HERE ################# 
    grad = None
    new_model = None
    
    grad_sprk = augmentedData.map(lambda x: ((np.sum(x[0]*W) - x[1])*x[0],1)) \
      .reduce(lambda x,y: (x[0]+y[0], x[1]+y[1]))
    grad = 2 * grad_sprk[0]/grad_sprk[1]
    new_model = W - learningRate * grad    
    ################## (END) YOUR CODE ################# 
   
    return new_model
```

```
# part b - OLS gradient descent function
def GradientDescent(trainRDD, testRDD, wInit, nSteps = 20, 
                    learningRate = 0.1, verbose = False):
    """
    Perform nSteps iterations of OLS gradient descent and 
    track loss on a test and train set. Return lists of
    test/train loss and the models themselves.
    """
    # initialize lists to track model performance
    train_history, test_history, model_history = [], [], []
    
    # perform n updates & compute test and train loss after each
    model = wInit
    for idx in range(nSteps): 
        
        ############## YOUR CODE HERE #############
        model = None
        training_loss = None
        test_loss = None
        #additional intermediate variables reused in each loop
        grad_sprk_q5b = None
        grad_q5b = None
        loss_sprk_q5_train = None
        loss_sprk_q5_test = None
        model_history_calc = None
        
        #Choosing prior weights
        if model_history == []:
            model_history_calc = wInit
        else:
            model_history_calc = model_history[-1:][0]
        
        #Block below mimics GDUpdate function
        #Model is established on Train data only
        #Note the embedded bias 'feature' of 1 at index 0
        grad_sprk_q5b = trainRDD.map(lambda x: (np.append([1.0], x[0]), x[1])) \
          .map(lambda x: ((np.sum(x[0]*model_history_calc) - x[1])*x[0],1)) \
          .reduce(lambda x,y: (x[0]+y[0], x[1]+y[1]))
        grad_q5b = 2 * grad_sprk_q5b[0]/grad_sprk_q5b[1]
        model = (model_history_calc - learningRate * grad_q5b).tolist()     

        #Block below mimics OLSLoss for trainRDD
        loss_sprk_q5_train = trainRDD.map(lambda x: (np.append([1.0], x[0]), x[1])) \
          .map(lambda x: ((np.sum(x[0]*model_history_calc) - x[1])**2,1)) \
          .reduce(lambda x,y: (x[0]+y[0], x[1]+y[1]))
        training_loss = loss_sprk_q5_train[0]/loss_sprk_q5_train[1]
        
        #Block below mimics OLSLoss for testRDD
        loss_sprk_q5_test = testRDD.map(lambda x: (np.append([1.0], x[0]), x[1])) \
          .map(lambda x: ((np.sum(x[0]*model_history_calc) - x[1])**2,1)) \
          .reduce(lambda x,y: (x[0]+y[0], x[1]+y[1]))
        test_loss = loss_sprk_q5_test[0]/loss_sprk_q5_test[1]        
        
        
        ############## (END) YOUR CODE #############
        
        # keep track of test/train loss for plotting
        train_history.append(training_loss)
        test_history.append(test_loss)
        model_history.append(model)
        # console output if desired
        if verbose:
            print("----------")
            print(f"STEP: {idx+1}")
            print(f"training loss: {training_loss}")
            print(f"test loss: {test_loss}")
            print(f"Model: {[np.round(w,3) for w in model]}")
    return train_history, test_history, model_history
```

```
# part b - helper function to emit residuals (RUN THIS CELL AS IS)
def kResiduals(dataPoint, models, splitNum):
    """
    Compute the (squared) residuals for a data point given k different models.
    Note that points from the k-th split are part of the test set for model number k
    and part of the training set for all other models. We'll emit a key to track this.
    Args:
        dataPoint - tuple of (features_array, y)
        models    - list of arrays representing model weights (bias at index 0)
    Returns:
        (stringFormattedKey, squared_error)
    """
    # augment the data point with a bias term at index 0
    X = np.append([1.0], dataPoint[0])
    y = dataPoint[1]
    # emit squared residuals for each model
    for modelNum, W in enumerate(models):
        if modelNum == splitNum:
            yield(f"{modelNum}-test", (W.dot(X) - y)**2)
        else:
            yield(f"{modelNum}-train", (W.dot(X) - y)**2)
```

```
# part c - fill in the missing code below
def CVLoss(dataSplits, models):
    """
    Compute the k-fold cross-validated test and train loss.
    Args:
        dataSplits - list of RDDs corresponding to the k test splits.
        models     - list of k arrays representing model weights (bias at index 0)
    Returns: 
        tuple of floats: (training_loss, test_loss)
    """
    # compute k residuals for each dataPoint (one for each model)
    partialLossRDD = sc.parallelize([])
    for splitNum, splitRDD in enumerate(dataSplits):
        residuals = splitRDD.flatMap(lambda x: kResiduals(x, models, splitNum))
        partialLossRDD = sc.union([partialLossRDD, residuals])
    
    ################ YOUR CODE HERE #################        
    loss = None
    
    countsByKey = sc.broadcast(partialLossRDD.countByKey())
    loss = partialLossRDD.reduceByKey(lambda x,y:x+y) \
      .map(lambda x: (x[0], x[1]/countsByKey.value[x[0]])) \
      .collect()
    
    ################ (END) YOUR CODE ################# 
    
    test_loss = np.mean([x[1] for x in loss if x[0].split('-')[1] == 'test'])
    training_loss = np.mean([x[1] for x in loss if x[0].split('-')[1] == 'train'])
    return training_loss, test_loss
```

```
# part d - helper function RUN THIS CELL AS IS
def partialGradients(splitNum, dataPoint, models):
    """
    Emit partial gradient for this data point for each model.
    NOTE: a data point from split-number k is in the test set for 
    model-k so we don't compute a partial gradient for that model.
    """
    # augment the data point
    X = np.append([1.0], dataPoint[0])
    y = dataPoint[1]
    # emit partial gradients for each model with a counter for averaging later
    for modelNum, W in enumerate(models):
        if modelNum != splitNum:
            yield (modelNum, [(W.dot(X) - y)*X, 1])
```

```
# part d - perform GD updates for all k models (FILL IN MISSING CODE BELOW)
def CVUpdate(dataSplits, models, learningRate = 0.1):
    """
    Compute gradients for k models given k corresponding dataSplits.
    NOTE: the training set for model-k is all records EXCEPT those in the k-th split.
    """
    # compute partial gradient k-1 times for each fold
    partialsRDD = sc.parallelize([])
    for splitNum, splitRDD in enumerate(dataSplits):
        thisFoldPartialGrads = splitRDD.flatMap(lambda x: partialGradients(splitNum, x, models))
        partialsRDD = sc.union([partialsRDD, thisFoldPartialGrads])

    # compute gradients by taking the average partialGrad for each fold
    gradients = partialsRDD.reduceByKey(lambda a, b: (a[0] + b[0], a[1] + b[1]))\
                           .mapValues(lambda x: x[0]/x[1])\
                           .map(lambda x: x[1])\
                           .collect()
    
    # update all k models & return them in a list
    new_models = []
    for W, grad in zip(models, gradients):
        ############# YOUR CODE HERE ############
        new_models.append(W - learningRate * grad) 
        ############# (END) YOUR CODE ###########
    return new_models
```

```
# part e - RUN THIS CELL AS IS
def GradientDescent_withCV(dataSplits, wInit, learningRate=0.1, nSteps = 5, verbose = False):
    """
    Train k models in parallel and track cross validated test/train loss.
    Returns:
        train_hist - (list) of floats
        test_hist - (list) of floats
        model_hist - (list) of arrays representing model coefficients (bias at index 0)
    """
    # broadcast initial models (one for each fold)
    bModels = sc.broadcast([wInit] * len(dataSplits))
    
    
    # initialize lists to track performance
    train_loss_0, test_loss_0 = CVLoss(dataSplits, bModels.value)
    train_hist, test_hist, model_hist = [train_loss_0], [test_loss_0], [wInit]
    
    # perform k gradient updates at a time (one for each fold)
    start = time.time()
    for step in range(nSteps):
        new_models = CVUpdate(dataSplits, bModels.value, learningRate)
           
        bModels = sc.broadcast(new_models)

        # log progress
        train_loss, test_loss = CVLoss(dataSplits, bModels.value)
        train_hist.append(train_loss)
        test_hist.append(test_loss)
        model_hist.append(new_models[0])
        
        if verbose:
            print("-------------------")
            print(f"STEP {step}: ")
            print(f"model 1: {[round(w,4) for w in new_models[0]]}")
            print(f" train loss: {round(train_loss,4)}")
            print(f" test loss: {round(test_loss,4)}")
            
    print(f"\n... trained {nSteps} iterations in {time.time() - start} seconds")
    return train_hist, test_hist, model_hist
```

# Question 7: Regularization.

Our goal, as always, is to build a linear model that will extend well to unseen data. Chosing the right combination of features to optimize generalizability can be extremely computationally costly given that there are \\(2^{p}\\) potential models that can be built from \\(p\\) features. Traditional methods like forward selection would involve iteratively testing these options to asses which combinations of features achieve a statistically significant prediction.

Ridge Regression and Lasso Regression are two popular alternatives to OLS, which enable us to train generalizable models without the trouble of forward selection and/or manual feature selection.  Both methods take advantage of the bias-variance tradeoff by _shrinking_ the model coefficients towards 0 which reduces the variance of our model with little increase in bias. In practice this 'shrinkage' is achieved by adding a penalty (a.k.a. 'regularization') term to the means squared error loss function.
```
# part c - gradient descent with regularization
def GDUpdate_wReg(dataRDD, W, learningRate = 0.1, regType = None, regParam = 0.1):
    """
    Perform one gradient descent step/update with ridge or lasso regularization.
    Args:
        dataRDD - tuple of (features_array, y)
        W       - (array) model coefficients with bias at index 0
        learningRate - (float) defaults to 0.1
        regType - (str) 'ridge' or 'lasso', defaults to None
        regParam - (float) regularization term coefficient
    Returns:
        model   - (array) updated coefficients, bias still at index 0
    """
    # augmented data
    augmentedData = dataRDD.map(lambda x: (np.append([1.0], x[0]), x[1]))
    
    new_model = None
    #################### YOUR CODE HERE ###################
    grad_root = augmentedData.map(lambda x: ((np.sum(x[0]*W) - x[1])*x[0],1)) \
      .reduce(lambda x,y: (x[0]+y[0], x[1]+y[1]))
    grad_intermed = 2*grad_root[0]/grad_root[1]
    
    if regType == 'ridge':
      grad = grad_intermed + 2*regParam*(np.insert(W[1:],0,0))
      
    elif regType == 'lasso':
      grad = grad_intermed + np.sign(np.insert(W[1:],0,0))*regParam
                               
    else: 
      grad = grad_intermed
                                                                                                                      
    new_model = W - learningRate * grad                          
    
    ################## (END) YOUR CODE ####################
    return new_model
```

```
# part d - ridge/lasso gradient descent function
def GradientDescent_wReg(trainRDD, testRDD, wInit, nSteps = 20, learningRate = 0.1,
                         regType = None, regParam = 0.1, verbose = False):
    """
    Perform nSteps iterations of regularized gradient descent and 
    track loss on a test and train set. Return lists of
    test/train loss and the models themselves.
    """
    # initialize lists to track model performance
    train_history, test_history, model_history = [], [], []
    
    # perform n updates & compute test and train loss after each
    model = wInit
    for idx in range(nSteps):  
        # update the model
        model = GDUpdate_wReg(trainRDD, model, learningRate, regType, regParam)
        
        # keep track of test/train loss for plotting
        train_history.append(OLSLoss(trainRDD, model))
        test_history.append(OLSLoss(testRDD, model))
        model_history.append(model)
        
        # console output if desired
        if verbose:
            print("----------")
            print(f"STEP: {idx+1}")
            print(f"training loss: {training_loss}")
            print(f"test loss: {test_loss}")
            print(f"Model: {[round(w,3) for w in model]}")
    return train_history, test_history, model_history
```

```
# part c - helper function (RUN THIS CELL AS IS)
def get_residuals(dataRDD, model):
    """
    Return a collected list of tuples (residual, quality_score)
    """
    augmentedData = dataRDD.map(lambda x: (np.append([1.0], x[0]), x[1]))
    residuals = augmentedData.map(lambda x: (x[1] - model.dot(x[0]), x[1]))
    return residuals.collect()
```

```
# part d - plotting function (RUN THIS CELL AS IS)
def plotCoeffs(models, featureNames, title):
    """
    Helper Function to show how coefficients change as we train.
    """
    fig, ax = plt.subplots(figsize = (15,8))
    X = list(range(len(models)))
    for data, name in zip(models.T, featureNames):
        if name == "Bias":
            continue
        ax.plot(X, data, label=name)
    ax.plot(X,[0]*len(X), 'k--')
    plt.title(title)
    plt.legend()
    display(plt.show())
```
